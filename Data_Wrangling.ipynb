```python
import pandas as pd

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"
df = pd.read_csv(file_path)
```

```python
#Finding duplicates
#In this section you will identify duplicate values in the dataset.
#Find how many duplicate rows exist in the dataframe.
print(df.duplicated().sum())
```

```python
print(df['Respondent'].duplicated().sum())
```

```python
#Removing duplicates
#Remove the duplicate rows from the dataframe.
df.drop_duplicates(inplace = True)
```

```python
df['Respondent'].value_counts()
```

```python
print(df.duplicated().value_counts())
```

```python
#Finding Missing values
df.isnull().any()
```

```python
df['Country'].isna().value_counts()
```

```python
#Find out how many rows are missing in the column 'WorkLoc'
df['WorkLoc'].isnull().sum()
```

```python
#Imputing missing values
#Find the value counts for the column WorkLoc.
df['WorkLoc'].value_counts()
```

```python
df['Employment'].value_counts()
```

```python
df['UndergradMajor'].value_counts()
```

```python
#dentify the value that is most frequent (majority) in the WorkLoc column.
df['WorkLoc'].value_counts().idxmax()
```

```python
#Impute (replace) all the empty rows in the column WorkLoc with the value that you have identified as majority.
df['WorkLoc'].fillna('Office', inplace=True)
```

```python
#Verify if imputing was successful.
df['WorkLoc'].isnull().sum()
```

```python
#Normalizing data
#List out the various categories in the column 'CompFreq'
df['CompFreq'].unique()
```

```python
df['CompFreq'].value_counts()
```

```python
#Create a new column named 'NormalizedAnnualCompensation'.
annComp=[]
def normalizedAnnualCompensation():
    for x,y in zip(df['CompFreq'], df['CompTotal']):
        if x=='Monthly':
            annComp.append(y*12)
        elif x=='Weekly':
            annComp.append(y*52)
        else:
            annComp.append(y)

normalizedAnnualCompensation()

df['NormalizedAnnualCompensation']=annComp
df[['NormalizedAnnualCompensation']]
```
